```python
import requests
from bs4 import BeautifulSoup
import pandas as pd
from selenium import webdriver
import pandas as pd
import time
from selenium.webdriver.support.ui import WebDriverWait
from selenium.webdriver.common.by import By
from selenium.webdriver.support import expected_conditions as EC

# Send an HTTP request to retrieve the HTML content of a webpage
url = 'https://finance.yahoo.com/quote/AAPL/history?p=AAPL'

driver = webdriver.Chrome()
driver.get(url)

try:
    popup_close_button = WebDriverWait(driver, 10).until(
        EC.element_to_be_clickable((By.XPATH, '//*[@id="myLightboxContainer"]/section/button[1]'))
    )
    popup_close_button.click()
except:
    print('Popup not found')

# Finding the address of the location
# response = BeautifulSoup(driver.page_source, 'html.parser')
# table_data = response.find('table', {'data-test': 'historical-prices'})

last_height = driver.execute_script('return document.body.scrollHeight')

# Scroll the webpage
driver.execute_script('window.scrollTo(0, {});'.format(last_height))
counter=0
while counter!=7:
    counter+=1
    # Wait for the page to load
    time.sleep(5)

    # Calculate the new height of the webpage
    new_height = driver.execute_script('return document.body.scrollHeight')
    print(last_height)
    print(new_height)
    # Check if we have reached the end of the table
#     if new_height == last_height:
#         break

    # Scroll the webpage to the new height
    driver.execute_script('window.scrollTo({}, {});'.format(last_height,last_height+new_height))

    # Update the last known height
    last_height = last_height+new_height



response = BeautifulSoup(driver.page_source, 'html.parser')
table_data = response.find('table', {'data-test': 'historical-prices'})

table_rows = table_data.find_all('tr')
header_cells = table_rows[0].find_all('th')

header = [cell.text for cell in header_cells]
print(header)
data = []
for row in table_rows[1:]:
    row_cells = row.find_all('td')
    row_data = [cell.text for cell in row_cells]
    data.append(row_data)
    
data=pd.DataFrame(data,columns=header)
print(data.head())
print(data.shape)

# Output the DataFrame to a CSV file
#df.to_csv('stock_data.csv', index=False)
```
